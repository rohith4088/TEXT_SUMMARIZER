```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate example data with a clear linear relationship
user_counts = np.array([50, 100, 150, 200, 250, 300, 350, 400, 450, 500])
prices = np.array([20,30,40,60,80,100,150,200,250,300])
revenue = user_counts * prices

# Create a linear regression model
model = LinearRegression()

# Reshape the data
user_counts = user_counts.reshape(-1, 1)

# Fit the model to the data
model.fit(user_counts, revenue)

# Predict revenue based on user counts
predicted_revenue = model.predict(user_counts)

# Plot the data and regression line
plt.scatter(user_counts, revenue, label='Actual Data')
plt.plot(user_counts, predicted_revenue, color='red', label='Linear Regression')
plt.xlabel('Number of Users')
plt.ylabel('Revenue')
plt.legend()
plt.show()

# Display the coefficients
print('Intercept (Price per User):', model.intercept_)
print('Coefficient (Number of Users):', model.coef_)
```
